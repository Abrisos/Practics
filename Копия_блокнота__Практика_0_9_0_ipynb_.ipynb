# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
def start():
  num = input("""Что будем делать? \n 1 -> Усыновить \n 2 -> Действия с питомцем
  3 -> Посмотреть состояние \n""")
  numHandler(num)


def numHandler(num):
  if num == '1':
    adopt()
  elif num == '2':
    pet_name = input('Введите имя питомца с которым хотите взаимодействовать: ')
    pet_interaction(pet_name)
  elif num == '3':
    pet_name = input('Введите имя питомца чье состояние хотите посмотреть: ')
    pet_status(pet_name)
  else:
    print('Вы ввели неправильный номер.')
  start()

def interactionHandler():
  num = input("""Что будем делать? \n 1 -> Покормить \n 2 -> Поиграть
  3 -> Уложить спать\n""")
  if num == '1':
    return 1
  if num == '2':
    return 2
  if num == '3':
    return 3
  return 0


def adopt():
  pet_name = input('Введите имя: ')
  pet_type = input('Введите тип: ')
  pet_age = input('Введите возраст: ')
  pets.append([pet_name, pet_type, pet_age, [100, 100, 100]])
  print('Питомец успешно усыновлен!')


def pet_interaction(pet_name):
  for name in pets:
    if pet_name == name[0]:
      num = interactionHandler()
      if num == 1:
        name[3][0] += 20
        if name[3][0] > 100:
          name[3][0] = 100
        name[3][1] -= 10
        name[3][2] -= 10
        print(f'Вы кормите питомца, теперь его голод: {name[3][0]}')
      elif num == 2:
        name[3][1] += 20
        if name[3][1] > 100:
          name[3][1] = 100
        name[3][0] -= 10
        name[3][2] -= 30
        print(f'Вы играйте с питомцем, теперь его счастье: {name[3][1]}')
      elif num == 3:
        name[3][2] += 30
        if name[3][2] > 100:
          name[3][2] = 100
        name[3][0] -= 30
        name[3][1] -= 10
        print(f'Питомец поспал, теперь его энергия: {name[3][2]}')
      else:
        print('Вы ввели неправильный номер.')

def pet_status(pet_name):
  for name in pets:
    if pet_name == name[0]:
      pet_hunger = name[3][0]
      pet_happy = name[3][1]
      pet_energy = name[3][2]
      if pet_hunger > 0 and pet_happy > 0 and pet_energy > 0:
        print(f'Голод: {pet_hunger}\nСчастье: {pet_happy}\nЭнергия: {pet_energy}')
      else:
        print('Питомец умер =(')
        pets.remove(name)


pets = []
print('Симулятор виртуального питомца')
start()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
from random import randint


def start():
  if not knight:
    print('Создайте рыцаря')
    knight_create()
  if not dragon:
    print('Создайте дракона')
    dragon_create()
  print('Игра началась.')
  knight_move()


def knight_create():
  name = input('Введите имя персонажа: ')
  armor = input('Введите информацию о броне персонажа: ')
  weapon = input('Введите информацию о оружии персонажа: ')
  health = randint(50,100)
  damage = randint(10,20)
  knight.extend([name, armor, weapon, health, damage])


def dragon_create():
  name = input('Введите имя дракона: ')
  health = randint(50,100)
  damage = randint(10,20)
  dragon.extend([name, health, damage])


def knight_move():
  print('Перед вами 3 двери, в какую пойдем?')
  door_num = int(input('\n1 -> Левая\n2 -> Средняя\n3 -> Правая\n'))
  if door_num == 1:
    if room_visiting[0] != 1:
      room_left()
    else:
      print('Вы уже там были')
      knight_move()
  if door_num == 2:
    room_center()
  if door_num == 3:
    if room_visiting[1] != 1:
      room_right()
    else:
      print('Вы уже там были')
      knight_move()


def room_left():
  print('Вы заходите в левую дверь и видите кота')
  move_num = int(input('1 -> Подойти к нему\n2 -> Уйти\n'))
  if move_num == 1:
    if randint(1,2) == 1:
      bonus_damage = randint(3,5)
      print(f'Кот дал вам дополнительные {bonus_damage} урона')
      knight[4] += bonus_damage
    else:
      forfeit_damage = randint(3,5)
      print(f'Кот забрал у вас {forfeit_damage} урона')
      knight[4] -= forfeit_damage
    room_visiting[0] += 1
  knight_move()


def room_right():
  print('Вы заходите в левую дверь и видите собаку')
  move_num = int(input('1 -> Подойти к ней\n2 -> Уйти\n'))
  if move_num == 1:
    if randint(1, 2) == 1:
      bonus_health = randint(10, 20)
      print(f'Собака дала вам {bonus_health} здоровья')
      knight[3] += bonus_health
    else:
      forfeit_health = randint(10, 20)
      print(f'Собака забрала у вас {forfeit_health} здоровья')
      knight[3] -= forfeit_health
    room_visiting[1] += 1
  knight_move()


def room_center():
  print(f'Вы заходите в дверь и видите дракона по имени {dragon[0]}!')
  print(f'Он говорит: Ого {knight[0]} у тебя что {knight[1]} броня '
        f'еще и {knight[2]} меч')
  print(f'Вы кричите ему: Давай сразимся {dragon[0]}!')
  while True:
    if knight_hit() == 1:
      print(f'Вы победили!\nУ вас осталось {knight[3]} здоровья!')
      break
    if dragon_hit() == 1:
      print(f'Вы проиграли!\nУ дракона осталось {dragon[1]} здоровья!')
      break


def knight_hit():
  hit = knight[4] + randint(-3, 3)
  input('Введите что-нибудь чтобы ударить!')
  print(f'Вы нанесли {hit} урона!')
  dragon[1] -= hit
  if dragon[1] > 0:
    print(f'У дракона осталось {dragon[1]} здоровья!\n')
    return 0
  return 1


def dragon_hit():
  hit = dragon[2] + randint(-4, 4)
  print(f'Дракон нанес вам {hit} урона!')
  knight[3] -= hit
  if knight[3] > 0:
    print(f'У вас осталось {knight[3]} здоровья!\n')
    return 0
  return 1


room_visiting = [0,0] # Посещения левой и правой комнаты
knight = []
dragon = []
print('Игра рыцарь и дракон')
start()
````
